# Theoretical Expression for Transmission Coefficeint

## General Description

The normalized propagating contant for travelling electron with energy $E$ in $z$ direction inside the step barrier potential profile $V(z)$ is given below. 
\begin{equation}
\psi(z) =
\begin{cases}
    A \sqrt{\frac{m_1}{\hbar k_1}} e^{ik_1z} + B \sqrt{\frac{m_1}{\hbar k_1}} e^{-ik_1z},& z< 0  \\
    C \sqrt{\frac{m_2}{\hbar k_2}} e^{ik_2z} + D \sqrt{\frac{m_2}{\hbar k_2}} e^{-ik_2z},              & z > 0
\end{cases}
\end{equation}

where, $m_1$ is the effective mass at the left side of the barrier and $k_1 = \sqrt{\frac{2m_1(E - V(z<0))}{\hbar^2}}$ , $m_2$ is the effective mass at the right side of the barrier and $k_2 = \sqrt{\frac{2m_2(E - V(z>0))}{\hbar^2}}$ and potential profile,
\begin{equation}
V(z) =
\begin{cases}
    0 ,&  z< 0  \\
   V_0              & z > 0
\end{cases}
\end{equation}

The value and slope of the wave function must be continuous everywhere. Therefore, we get the following eqautions at $z=0$.

\begin{equation}
\sqrt{\frac{m_1}{\hbar k_1}}(A+B) = \sqrt{\frac{m_2}{\hbar k_2}} (C+D),
\sqrt{\frac{m_1}{\hbar k_1}}k_1(A-B) = \sqrt{\frac{m_2}{\hbar k_2}}k_2(C-D)
\end{equation}

Next, we need to write down the above eqautions into matrix equation,

\begin{equation}
\begin{bmatrix}
    \sqrt{\frac{m_1}{\hbar k_1}}  &  \sqrt{\frac{m_1}{\hbar k_1}}      \\
    k_1 \sqrt{\frac{m_1}{\hbar k_1}}  &  -k_1 \sqrt{\frac{m_1}{\hbar k_1}}    
\end{bmatrix} 
\begin{bmatrix}
    A      \\
    B      
\end{bmatrix} 
=
\begin{bmatrix}
    \sqrt{\frac{m_2}{\hbar k_2}}  &  \sqrt{\frac{m_2}{\hbar k_2}}      \\
    k_2 \sqrt{\frac{m_2}{\hbar k_2}}  &  -k_2 \sqrt{\frac{m_2}{\hbar k_2}}    
\end{bmatrix} 
\begin{bmatrix}
    C      \\
    D      
\end{bmatrix}
\end{equation}

Finally we can represent $C,D$ in terms of $A,B$.
\begin{equation}
\sqrt{\frac{m_1}{4 m_2 k_2 k_1}}
\begin{bmatrix}
    (k_2+k_1)  &  (k_2-k_1)      \\
   (k_2-k_1)  &  (k_2+k_1)    
\end{bmatrix} 
\begin{bmatrix}
    A      \\
    B      
\end{bmatrix} 
=
\begin{bmatrix}
    C      \\
    D      
\end{bmatrix}
\end{equation}

Now using the definition of **Transfer Matrix Theory**, we get,

\begin{equation}
T(0)=
\sqrt{\frac{m_1}{4 m_2 k_2 k_1}}
\begin{bmatrix}
    (k_2+k_1)  &  (k_2-k_1)      \\
   (k_2-k_1)  &  (k_2+k_1)    
\end{bmatrix}
\end{equation}

If we shift the barrier to the $z = +d$ location we need to add extra phase factor to the solution and the *transfer matrix* has its new form,
\begin{equation}
T(d)=
\begin{bmatrix}
     e^{-ik_2d} & 0       \\
     0 &  e^{ik_2d}   
\end{bmatrix} 
T(0)
\begin{bmatrix}
     e^{ik_1d} & 0       \\
     0 &  e^{-ik_1d}   
\end{bmatrix} 
\end{equation}

In order to find the flux transmission coefficient $\mathbf{T}$. we need to find the **transfer matrix** for the whole system and then according to the construction of equation(1):
\begin{equation}
\mathbf{T} = \frac{{|C|}^2}{{|A|}^2}
\end{equation}

This value can be easily derived from the **transfer matrix**. If we consider incident part $A = 1$ , reflected part of the wave function $B = r$, transmitted part $C = t$ and wave coming from opposite direction in the barrier $D = 0$. We can deduce following results.
For, 
\begin{equation}
\begin{bmatrix}
    T_{11}  &  T_{12}      \\
    T_{21}  &  T_{22}    
\end{bmatrix} 
\begin{bmatrix}
    1      \\
    r      
\end{bmatrix} 
=
\begin{bmatrix}
    t      \\
    0      
\end{bmatrix}
\end{equation}
we have,
\begin{equation}
t = \frac{T_{11}T_{22} - T_{12}T_{21}}{T_{22}} , \ r = -\frac{T_{21}}{T_{22}}
\end{equation}


Furthermore for applying time symmetry condition in Schrodinger equation and the equal current for both side we obtain,

\begin{equation}
det(T) = T_{11}T_{22} - T_{12}T_{21} = 1 
\end{equation}
Thus,
\begin{equation}
t = \frac{1}{T_{22}} \ \text{and} \ \mathbf{T} = \frac{1}{|T_{22}|^2}
\end{equation}

## Problem Specifications

An asymmetrical rectangular barrier is used to represent a 1-D transistor, with Region 1 ($U_1$) being the source, Region 2 ($U_2$ being the channel, and Region 3 ($U_3$) being the drain. For a particular example,let the energies in eV to be $0, 0.1$ and $-0.5$ for $U_1, U_2$ and $U_3$, respectively. Also the effective masses are $m_1 = m_3 = 0.3m_0$ and $m_2 = 0.1m$ . The channel is $10 \ nm$ long and the task is to solve for the transmission probability $\mathbf{T}(E)$ for energies up to about $10k_BT$ above $U_1$ .

if we translate it into **atomic unit** the formuals get changed where $1 \ eV = 0.0367493 \ hartree$ and $ 1 \ nm = 18.897161646321 \ bohr $ and mass of free electron $m_0 = 1 \ a.u.$, Planck's reduced constant , $\hbar = 1 \ a.u.$

V(z) =

## Theoretical Calculation for Given System

Just for simplicity, Let's assume the barrier expand from $-\frac{a}{2} \leq z \leq \frac{a}{2}$. Now the *T* matrix from region I to region II is given below.

\begin{equation}
T^{(21)}=
\begin{bmatrix}
     e^{-ik_2(\frac{-a}{2})} & 0       \\
     0 &  e^{ik_2(\frac{-a}{2})}   
\end{bmatrix} 
\sqrt{\frac{m_1}{4 m_2 k_2 k_1}}
\begin{bmatrix}
    (k_2+k_1)  &  (k_2-k_1)      \\
   (k_2-k_1)  &  (k_2+k_1)    
\end{bmatrix}
\begin{bmatrix}
     e^{ik_1(\frac{-a}{2})} & 0       \\
     0 &  e^{-ik_1(\frac{-a}{2})}   
\end{bmatrix} 
\end{equation}

Here, $k_1 = \sqrt{\frac{2m_1(E - V_1)}{\hbar^2}}$ and $k_2 = \sqrt{\frac{2m_2(E - V_2)}{\hbar^2}}$. Similarly, *T* matrix for region II to region III we can write,

\begin{equation}
T^{(32)}=
\begin{bmatrix}
     e^{-ik_3(\frac{a}{2})} & 0       \\
     0 &  e^{ik_3(\frac{a}{2})}   
\end{bmatrix} 
\sqrt{\frac{m_2}{4 m_1 k_2 k_3}}
\begin{bmatrix}
    (k_3+k_2)  &  (k_3-k_2)      \\
   (k_3-k_2)  &  (k_3+k_2)    
\end{bmatrix}
\begin{bmatrix}
     e^{ik_2(\frac{a}{2})} & 0       \\
     0 &  e^{-ik_2(\frac{a}{2})}   
\end{bmatrix} 
\end{equation}

Here, $k_3 = \sqrt{\frac{2m_1(E - V_3)}{\hbar^2}}$. The complete transfer matrix for this system would be,
\begin{equation}
T^{(31)} = T^{(32)}T^{(21)}
\end{equation}

But our concern is only about $|T^{(31)}_{22}|^2$ which is needed to determine the transmission coefficient. Finally by evaluating the matrix $T^{(31)}$ we get,

\begin{equation}
T^{(31)}_{22} = \frac{e^{ik_2a}(k_2-k_1)(k_3-k_2)+ e^{-ik_2a}(k_2+k_1)(k_3+k_2)}{4\sqrt{k_1k_2^2k_3}}e^{i(k_1+k_3)\frac{a}{2}}
\end{equation}

\begin{equation}
\implies \mathbf{T} = \frac{1}{|T^{(31)}_{22}|^2} = \frac{4k_1k_2^2k_3}{(k_2k_3 + k_1k_2)^2cos^2{k_2a} + (k_2^2+k_1k_3)^2sin^2{k_2a}}
\end{equation}

# NEGF Calculation

# NEGF Formalism

To find the *transmission probability*, we are going to use Non-Equilibrium Green Function (NEGF) method. From NEGF theory Fisher-Lee relation is,
\begin{equation}
\mathbf{T}(E) = Tr[\Gamma_L(E)G(E)\Gamma_R(E)G^\dagger(E)]
\end{equation}

where, braodening matrix of left contact and right conatact are $\Gamma_L(E) = i[\Sigma_L(E)-\Sigma^\dagger_L(E)]$ and $\Gamma_R(E) = i[\Sigma_R(E)-\Sigma^\dagger_R(E)]$ respectively,  $\Sigma_L(E)$ and $\Sigma_R(E)$ are the self energy matrix for left contact respectively. Furthermore, the Green function is $G(E) = {(E-i0^+I-H-\Sigma_L(E)-\Sigma_R(E))}^{-1}$ according to **\cite{1}** but according to **\cite{2}** formula for Green function is $G(E) = {(E-i0^+I-H+\Sigma_L(E)+\Sigma_R(E))}^{-1}$, here $H$ is the Hamiltonian of the system.

In **\cite{2}** authors suggest the following formulas for 1D constant potential calculation. For this 1D system the self energy function for left contact according to **\cite{2}** is given below:

\begin{equation}
\Sigma_L(E) = -t_L^2g_L(E)\delta_{11}
\end{equation}

where, $t_L = \frac{\hbar^2}{2m_La^2}$ , $g_L = \frac{2t_L + V_L - E \pm \sqrt{ (2t_L + V_L - E)^2 - 4t_L^2}}{-2t_L^2}$ and $\delta_{11}$ is Kronecker delta function. Similarly $\Sigma_R(E)$ can be defined by changing index $L$ to $R$ and $\delta_{11} \to \delta_{NN}$. Consequently, the formula for self energy is ($L$ for left and $R$ for right):

\begin{equation}
\Sigma_{L \ / \ R} = \frac{2t_{L \ / \ R} + V_{L \ / \ R} - E \pm \sqrt{ (2t_{L\ / \ R} + V_{L \ / \ R} - E)^2 - 4t_{L \ / \ R}^2}}{2}
\end{equation}

On the other hand, Supriyo Datta **\cite{1}** used slightly different apporach while considering one-dimensional NEGF approach. There author defined the formula for self energy:
\begin{equation}
\Sigma (E) = - t e^{ika} 
\end{equation}

Here $t = \frac{\hbar^2}{2ma^2} $ and the value of $k  a$ can be found from the energy formula for open system $E = V + 2 t \big( 1 - cos({ka}) \big) $ . We have dropped the subscripts for simplicity. Therefore, 

\begin{equation}
ka = 1 - cos^{-1}\big(\frac{E - V}{2t}\big)
\end{equation}

## Equivalence between Two Formulation

In this subsection we are going to show the both formalism we have mentioned before. We will be using equation (23) to derive euqation (21), or we might say that we will demonstarte that both staisfy same quardatic equation. 

From equation (23) it is easy to observe that ,
\begin{equation}
cos(ka) = \frac{2t + V - E}{2t}
\end{equation}

Therefore, we can rewrite the formula for equation (22):
\begin{equation}
\Sigma(E) = -t ( cos(ka) + i sin(ka))= -t\frac{2t + V - E}{2t} - it\frac{\sqrt{4t^2 - (2t + V - E)^2}}{2t}
\end{equation}

Finally we have the same form for self energy as in the equation (21).

\begin{equation}
\Sigma(E) =  -\frac{2t + V - E \ + \ it\sqrt{4t^2 - (2t + V - E)^2}}{2}
\end{equation}

The negative sign is due to difference in definition of self energy in two references. If $ \sqrt{4t^2 - (2t + V - E)^2 }$ is positive we get a complex solution for the self energy, we can get same solution from the equation (21) when the discriminator is negative. Similarly, for negative value of  $ \sqrt{4t^2 - (2t + V - E)^2} $ we would obtain the equivalent result from equation (21) with negative sign. Therefore, we can conclude that, self energy $\Sigma$ in equation (26) must satisfy, (using definition from **cite2**)
\begin{equation}
\Sigma = \frac{t^2}{E - 2t - V - \Sigma}
\end{equation}

It is possible to achieve other sets of solution (given in equation (21)) of equation (27) by changing the sign of $k$ in the self energy formula given in equation (22). Physically it means the same energy but with negative momentum states. Furthermore, the complex exponential is an eigen-function for semi-infinite Hamiltonian matrix which is the primary motivation for transfer matrix method. NEGF in one-dimension is just different but elegant perspective of investagating transport in nanostructures with broader reach.

In this following code snippet we are going to find transmission probability for the given system.

```matlab
% all computation is done in atomic unit
format long
m1 = 0.3; m3 = 0.3;    % eff. mass in left and right contact
m2 = 0.1;              % eff. mass in channel region
a = 1;                 % sampling distance in (bohr)

N1 = 105;              % # of data points in source region
N2 = 190;              % # of data points in channel region 
N3 = 105;              % # of data points in drain region
N = N1 + N2 + N3;      % # of data points total

t_L = 1/2/m1/a^2;      % in atomic unit
t_R = 1/2/m3/a^2;      % in atomic unit

V1 = 0;                % pot. in left contact
V2 = 0.00367493;       % pot. in channel
V3 = -0.01837465;      % pot. im right contact

V = [V1*ones(1,N1),V2*ones(1,N2),V3*ones(1,N3)];
                       % potential profile
tau = 2*[m1*ones(1,N1),m2*ones(1,N2),m3*ones(1,N3)]*a^2;
                       % multiplier for Hamiltonian
H = diag(1./tau)*toeplitz([2 -1 zeros(1,N-2)]) + diag(V);
                       % Complete Hamiltonian
H = sparse(H);

Energy = linspace(0,0.5,500) * 0.0367493; % hartree conv.
                       % Energy sampling from 0 to 0.5 eV
T = [];                % transmission coefficient

sigma_L = sparse(zeros(N));
                       % self energy for left contact
sigma_R = sparse(zeros(N));
                       % self energy for right contact
zplus = 1i*1e-12;      % leakage part in Hamiltonian

for E = Energy
    % commented part will simulate PhysRevB.69.165301
    %EL = 2*t_L + V1 - E;
    %sigma_L(1,1) = (EL+sqrt(EL^2-4*t_L^2))/2;
    EL = 1 - (E+zplus-V1)/2/t_L; ka = acos(EL);
    sigma_L(1,1) = -t_L*exp(1i*ka);
    %ER = 2*t_R + V3 - E;
    %sigma_R(N,N) = (ER+sqrt(ER^2-4*t_R^2))/2;
    ER = 1 - (E+zplus-V3)/2/t_R; kb = acos(ER);
    sigma_R(N,N) = -t_R*exp(1i*kb);
    
    Gamma_L = 1i*(sigma_L-sigma_L');
    Gamma_R = 1i*(sigma_R-sigma_R');
    % G = inv((E+zplus)*speye(N)-H+sigma_L+sigma_R);
    G = inv((E+zplus)*speye(N)-H-sigma_L-sigma_R);
    T = [T , real(trace(Gamma_L*G*Gamma_R*G'))];
end
figure
hold on
plot(Energy/0.0367493,T,'linewidth',3)
width = 190;
k1 = sqrt(2* m1 * (Energy-V1));
k2 = sqrt(2* m2 * (Energy-V2));
k3 = sqrt(2* m1 * (Energy-V3));
T_theory = 4*k1.*(k2.^2).*k3./((k2.*k3 + k1.*k2).^2.* ...
    (cos(k2*width)).^2+(k2.*k2 + k1.*k3).^2.*(sin(k2*width)).^2);
plot(Energy/0.0367493,real(T_theory),'-*')
plot([V2, V2]/0.0367493,[0,1],'k', 'linewidth',2)
grid on
legend('NEGF Calculation','Theoretical Formula','location','southeast')
text(0.11, 0.95, 'Barrier Height')
ylabel('Transmission Probability T(E)', 'fontweight','bold')
xlabel('Energy(eV)','fontweight','bold')
title('Theroretical and NEGF calculation of T(E)')
```
